##### import tips dataset.

```python
### code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips=pd.read_csv(r"C:\Users\DELL\Untitled Folder\IBM_Python_exercises\tips.csv")
tips
```

##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

```python
sns.set_palette('bright')
sns.set_style('whitegrid')

sns.barplot(data=tips, x='sex', y='total_bill', ci=None)

plt.title('Total Bill by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Bill')

plt.show()
```

###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

```python
sns.boxplot(y='total_bill',x='day',hue='smoker',data=tips)
plt.title('Total Bill by Day and Smoker Status')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here

sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)

plt.title('Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here 
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', jitter=True)

plt.title('Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

##### 7. Write a Python program to display density plot of all numerical columns from the dataframe and write the intution about plot.

```python
### code here
fig, axs = plt.subplots(ncols=3, nrows=1, figsize=(15, 5))

sns.kdeplot(data=tips, x='total_bill', ax=axs[0])
sns.kdeplot(data=tips, x='tip', ax=axs[1])
sns.kdeplot(data=tips, x='size', ax=axs[2])


plt.tight_layout()
plt.show()
```

##### 8. Write a Python program to calculate average total bills by days and display top 4 days by plot.

```python
### code here
avg_total_bills = tips.groupby('day')['total_bill'].mean().sort_values(ascending=False)[:4]

# plot the top 4 days
sns.barplot(x=avg_total_bills.index, y=avg_total_bills.values, palette='Blues_d')
sns.despine()
sns.set_style('whitegrid')
sns.set_context('talk', font_scale=0.8)
plt.title('Top 4 Days with Highest Average Total Bills')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
```
