######1. load the iris.csv dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris=pd.read_csv(r"C:\Users\DELL\Untitled Folder\IBM_Python_exercises\Iris.csv")
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
sns.countplot(x='Species', data=iris, )
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here

species_count = iris['Species'].value_counts()

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(species_count.values, labels=species_count.index, autopct='%1.1f%%')
ax.set_title('Percentage of Iris Species')
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here



sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm', data=iris)
sns.jointplot(x='SepalLengthCm', y='PetalWidthCm', data=iris)
sns.jointplot(x='SepalWidthCm', y='PetalLengthCm', data=iris)
sns.jointplot(x='SepalWidthCm', y='PetalWidthCm', data=iris)
sns.jointplot(x='PetalLengthCm', y='PetalWidthCm', data=iris)


plt.tight_layout()
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',
                hue='Species', data=iris, )
 

plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
sns.boxplot(x=iris['SepalLengthCm'])
plt.show()
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(x=iris['SepalLengthCm'])
plt.show()
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(iris.drop(['Id'], axis = 1),hue='Species', height=2)
```

##### 9. Create a Heat Map

```python
### code here
sns.heatmap(iris.corr(method='pearson').drop(['Id'], 
           axis=1).drop(['Id'], axis=0),annot = True);
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
sns.set(style = 'darkgrid')
sns.distplot(iris['SepalLengthCm'])
plt.show()
```
