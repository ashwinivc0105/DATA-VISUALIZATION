```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
diamonds=pd.read_csv(r"C:\Users\DELL\Untitled Folder\IBM_Python_exercises\diamonds.csv")
diamonds.head()
```

#### Write a Python program to display all Univariate Plots from the dataframe.


1.   Numeric based plots
2.   Categorical based plots




```python

# Numeric-based plots
sns.displot(diamonds['carat'], kde=True)
plt.show()

sns.displot(diamonds['depth'], kde=True)
plt.show()

sns.displot(diamonds['table'], kde=True)
plt.show()

sns.displot(diamonds['price'], kde=True)
plt.show()

sns.displot(diamonds['x'], kde=True)
plt.show()

sns.displot(diamonds['y'], kde=True)
plt.show()

sns.displot(diamonds['z'], kde=True)
plt.show()
```

```python
# Categorical-based plots
sns.countplot(x='cut', data=diamonds)
plt.show()

sns.countplot(x='color', data=diamonds)
plt.show()

sns.countplot(x='clarity', data=diamonds)
plt.show()

sns.boxplot(x='cut', y='price', data=diamonds)
plt.show()

sns.boxplot(x='color', y='price', data=diamonds)
plt.show()

sns.boxplot(x='clarity', y='price', data=diamonds)
plt.show()
```

```python
#1.Histogram of diamond carat weight:
sns.histplot(data=diamonds, x='carat', kde=True)
plt.show()
```

```python
#2.Kernel density plot of diamond carat weight:
sns.kdeplot(data=diamonds, x='carat')
plt.show()
```

```python
#3.Box plot of diamond carat weight:
sns.boxplot(data=diamonds, y='carat')
plt.show()
```

```python
#4.Violin plot of diamond price by cut:
sns.violinplot(data=diamonds, x='cut', y='price')
plt.show()
```

```python
#6.Count plot of diamond color:
sns.countplot(data=diamonds, x='color')
plt.show()
```

```python
#7. Bar plot of diamond price by cut and color:
sns.barplot(data=diamonds, x='cut', y='price', hue='color')
plt.show()
```

```python
#8.Scatter plot of diamond price vs. carat weight:
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.show()
```

```python
#9. Line plot of average diamond price by carat weight:
avg_price_by_carat = diamonds.groupby('carat').mean().reset_index()
sns.lineplot(data=avg_price_by_carat, x='carat', y='price')
plt.show()
```

#### Write a Python program to display all Multivariate plots from the dataframe.

```python
# Pairplot for numeric columns
sns.pairplot(data=diamonds)
```

```python
# Heatmap for correlation between numeric columns
sns.heatmap(data=diamonds.corr(), cmap='coolwarm', annot=True)
```

```python

# Scatterplot for numeric to numeric with hue for categorical column
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut')
```

```python
# Boxplot for categorical to numeric with hue for categorical column
sns.boxplot(data=diamonds, x='cut', y='price', hue='color')
```

```python
# Barplot for categorical to numeric with hue for categorical column
sns.barplot(data=diamonds, x='cut', y='price', hue='color')
```

```python
# Violinplot for categorical to numeric with hue for categorical column
sns.violinplot(data=diamonds, x='cut', y='price', hue='color')
```

```python
# PairGrid for numeric to numeric with hue for categorical column
g = sns.PairGrid(diamonds, x_vars=["carat", "depth", "table"], y_vars=["price"], hue='cut')
g.map(sns.scatterplot)
g.add_legend()
```
