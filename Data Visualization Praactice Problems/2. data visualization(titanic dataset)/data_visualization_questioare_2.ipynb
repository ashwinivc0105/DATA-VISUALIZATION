##### import the titanic dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
titanic=pd.read_csv(r"C:\Users\DELL\Untitled Folder\IBM_Python_exercises\titanic.csv")
titanic
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
import seaborn as sns

titanic['Fare_Range'] = pd.qcut(titanic['fare'], 4)
sns.barplot(x ='Fare_Range', y ='survived',
data = titanic)
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here

sns.set_palette('bright')
sns.set_style('whitegrid')

class_fares = titanic.groupby('class')['fare'].mean().reset_index()

sns.barplot(data=class_fares, x='fare', y='class', orient='h')

plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
sns.set_palette('bright')
sns.set_style('whitegrid')

class_fares = titanic.groupby('class')['fare'].mean().reset_index()

colors = ['blue', 'green', 'red']

sns.barplot(data=class_fares, x='class', y='fare', palette=colors)

plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
### code here

class_fares = titanic.groupby('class')['fare'].sum()

plt.pie(class_fares, labels=class_fares.index, autopct='%1.1f%%')

plt.axis('equal')

plt.show()
```

######5. Write a Python program to display a count plot of all categorical columns.

```python
### code here

categorical_cols = ['sex', 'class', 'deck', 'embark_town', 'alone']

for col in categorical_cols:
    sns.countplot(data=titanic, x=col)
    plt.title(f"Count Plot of {col.capitalize()}")
    plt.show()
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
### code here

num_cols = ['age', 'fare']

for col in num_cols:
    sns.displot(data=titanic, x=col, kde=True)

plt.show()
```

###### 10. Write a Python program to display boxplot of age, pclass and hue = gender.
